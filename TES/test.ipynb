```python
import numpy as np
```

```python

# BOOLEAN STUFF 

# Imagine you have a list of numbers representing the "Categories" of items in a small dataset:
# categories = [0, 1, 2, 0, 3, 2, 1, 0]

# Question 1: The "Bouncer" Test
# Goal: Create a list of True and False (a mask) that identifies only the zeros.
```

```python

categories = np.array([0, 1, 2, 0, 3, 2, 1, 0])
mask = categories == 0
print(mask)
```

```python

# Question 2: The "Either/Or" (The | Operator)
# Goal: You want to find both category 0 and category 2.

mask = (categories == 0) | (categories == 2)
print(mask)
```

```python
# Question 3: "Opposite Day" (The ~ Operator)
# Goal: You have a mask that finds category 0. Now you want everything except category 0.
```

```python
mask = categories == 0
inverted_mask = ~mask
print(inverted_mask)
```

```python
# Question 4: The "Filter" (Applying the Mask)
# Goal: Use your mask to actually extract the data.

# Challenge: If data = ["Apple", "Banana", "Cherry", "Date"] and your mask = [True, False, True, False], what will data[mask] give you?
```

```python
data = np.array(["Apple", "Banana", "Cherry", "Date"])
mask = (data == "Apple") | (data == "Cherry")
print(data[~mask])
```

```python
# Question 5: The "Relabel" Logic
# Goal: You removed category 0. Now your labels are [1, 2, 3]. You need them to be [0, 1, 2].


y = np.array([1 , 2 ,3 ])
y = y - 1
y
```
