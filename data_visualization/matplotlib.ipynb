```python
import matplotlib.pyplot as plt
import matplotlib as mlb
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,5,11)
y = x **2
```

```python
plt.plot(x, y)
plt.title("X and Y function graph")
plt.xlabel("X Axis")
plt.show()
```

```python
#subplots
```

```python
plt.subplot(2,2,1)
plt.plot(x)
plt.subplot(2,2,2)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(x**2 , y**2)
plt.subplot(2,2,4)
plt.plot(y,x)
```

```python
#object oriented way
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1 , 0.1 , 1 , 1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.1 , 0.1 , 0.8 , 0.8])
axis2.plot(x,y)
```
