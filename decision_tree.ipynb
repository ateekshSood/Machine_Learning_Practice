```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
import numpy as np
import graphviz
```

```python
iris = load_iris()
x = iris.data[: , 2:]
y = iris.target
```

```python
tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(x,y)
```

```python
data = export_graphviz(
    tree_clf,
    feature_names=iris.feature_names[2:],
    class_names=iris.target_names,
    rounded = True,
    filled = True
)

graph = graphviz.Source(data)
graph
```

```python
y_pred = tree_clf.predict(x)
decision_boundary = tree_clf.decision_path(x)

decision_boundary.shape
```

```python
def plot_points(x , y):
    plt.scatter(x[: , 0] , x[: , 1] , c=y)
    
```

```python
def plot_decision(model , axes):
    x1 = np.linspace(axes[0] , axes[1] , 100)
    x2 = np.linspace(axes[2] , axes[3] , 100)
    x1_meshgrid , x2_meshgrid = np.meshgrid(x1 , x2)
    x_final_points = np.c_[x1_meshgrid.ravel() , x2_meshgrid.ravel()]
    y_pred = tree_clf.predict(x_final_points).reshape(x1_meshgrid.shape)
    plt.contourf(x1_meshgrid , x2_meshgrid , y_pred ,alpha = 0.3)
    plt.axis(axes)
    plt.xlabel("petal length" , fontsize = 14)
    plt.ylabel("petal width" , fontsize=14)
    plt.show()
    

plt.plot([2.45 , 2.45] , [0,3], "k-")
plt.plot([2.45 , 7.5] , [1.75 , 1.75] , "k--")
plt.title("max depth = 2" , fontsize = 32)
plot_points(x , y)
plot_decision(tree_clf , [0, 7.5, 0, 3])

```

```python
max_depth_ = 3
another_tree_clf = Pipeline([
    ("tree" , DecisionTreeClassifier(max_depth=max_depth_))
])

another_tree_clf.fit(x , y)
```

```python
tree_image = export_graphviz(
    another_tree_clf[0],
    feature_names=iris.feature_names[2:],
    class_names=iris.target_names,
    rounded=True,
    filled = True
)

graph = graphviz.Source(tree_image)
graph
```

```python
plot_points(x , y)
plt.plot([0 , 7.5] , [0.8 , 0.8] ,"k-", linewidth= 2 )
plt.plot([0 , 7.5] , [1.75 , 1.75] , "k--" , linewidth=2)
plt.plot([4.95 , 4.95] , [1.75 , 3] , "k:")
plt.plot([4.85 , 4.85] , [0 , 1.75] , "k:")
plt.text(0.5 , 0.90 , "Decision Boundary 1" )
plt.text(0.5 , 1.90 , "Decision Boundary 2")
plt.text(2.5 , 2.5 , "Decision Boundary 3")
plt.text(5 , 1.25 , "Decision Boundary 4")
plt.title("max depth = 3" , fontsize = 21)
plot_decision(another_tree_clf , [0, 7.5, 0, 3])
```

```python
#REGRESSION 

np.random.seed(42)
m = 200
X = np.random.rand(m, 1)
y = 4 * (X - 0.5) ** 2
y = y + np.random.randn(m, 1) / 10
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(max_depth=2)
tree_reg.fit(X , y)

regression_tree = export_graphviz(
    tree_reg,
    rounded=True,
    filled=True
) 

i_dont_have_variable_name_in_mind = graphviz.Source(regression_tree)
i_dont_have_variable_name_in_mind
```

```python
tree_reg = DecisionTreeRegressor(max_depth=3)
tree_reg.fit(X , y)

regression_tree = export_graphviz(
    tree_reg,
    rounded=True,
    filled=True
) 

i_dont_have_variable_name_in_mind = graphviz.Source(regression_tree)
i_dont_have_variable_name_in_mind
```

```python
def plot_regression(x , y , tree , axes):
    x1 = np.linspace(axes[0] , axes[1] , 500).reshape(-1,1)
    y_pred = tree.predict(x1)
    plt.plot(x , y , "b.")
    plt.plot(x1 , y_pred , "k-" , linewidth=2)
    plt.axis(axes)
    
plot_regression(X , y , tree_reg , [0, 1, -0.2, 1])
plt.plot([0.197 , 0.197] , [-0.2 , 1.2] , "r--")
plt.text(0.24 , 0.9 , "Decision Boundary 1")
plt.plot([0.092 , 0.092] , [-0.2 , 1] , "r--")
plt.text(0, 0.3 , "Decision Boundary 2")
plt.plot([0.772 , 0.772] , [-0.2 , 1] , "r--")
plt.text(0.8 , 0 , "Decision Boundary 3")
plt.title("max depth = 2" , fontsize=21)
plt.show() 
    
```
