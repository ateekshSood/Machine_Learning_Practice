```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = 2 * np.random.rand(100,1)
y = 4 + 3 * x + np.random.rand(100,1)
```

```python
x[0] , y[0]
```

```python
plt.plot(x , y , "b.")
plt.axis(True)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# the normal equatin that minimizes the cost function 

#theta = (x transpose x) inv x transpose y
```

```python
x_b = np.c_[np.ones((100 , 1)) , x]
theta_best = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(y)

#we concatinated the arrays cuz we had to find the best theta for our intercept too 4 in this case
```

```python
theta_best
```

```python
x_new = np.array([[0] , [2]])
```

```python
x_new_b = np.c_[np.ones((2,1)) , x_new]
```

```python
y_predict = x_new_b.dot(theta_best)
```

```python
y_predict
```

```python
plt.plot(x_new , y_predict , "r-")
plt.plot(x , y , "b.")
plt.axis([0,2,0,15])
plt.show()
```

```python
m = 100
x = 6 * np.random.rand(m , 1) - 3
y = 0.5 * x**2 + x + 2 + np.random.rand(m , 1)
```

```python
plt.plot(x , y , "b.")
plt.xlabel("X")
plt.ylabel("y")
plt.axis([-3 , 3 , 0 , 10])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

another_lin_reg = LinearRegression()
another_lin_reg.fit(x , y)
```

```python
y_pred = another_lin_reg.predict(np.linspace(-3 , 3 , 100).reshape(100 ,1 ))
```

```python
plt.plot(x , y , "b.")
plt.plot(y_pred , "r-")
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree = 2 , include_bias = False)
x_poly_featueres = poly_features.fit_transform(x)
```

```python
x[0]
```

```python
x_poly_featueres[0]
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x_poly_featueres , y)
```

```python
lin_reg.fit_intercept , lin_reg.coef_
```

```python
x_new = np.linspace(-3 , 3 , 100).reshape(100 ,1)
x_new_poly = poly_features.fit_transform(x_new)
y_pred = lin_reg.predict(x_new_poly)
```

```python
plt.scatter(x , y)
plt.plot(x_new , y_pred ,"r-")
plt.axis([-3 , 3 , 0 , 10])
plt.xlabel("X values" , fontsize = 19)
plt.ylabel("y" , rotation = 0 , fontsize = 19)
plt.show()
```
