```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-6 , 6 , 500)
y = 1/( 1 + np.exp(-x))

plt.plot(x , y , "r-" , label=r"$\frac{1}{1+exp^{-x}}$")
plt.legend(loc="upper left" , fontsize=13)
plt.axis([-6 , 6 , 0 , 1])
plt.plot([0 , 0] , [ 0, 1] , "k")
plt.plot([-6,6] , [0.5,0.5] , "g--")
plt.title("sigmund function")

plt.show()

```
