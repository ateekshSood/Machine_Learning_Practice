```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import add_dummy_feature
from sklearn.linear_model import LinearRegression
```

```python
#linear reg equation : y_pred = theta.T.dot(X) + bias
```

```python
#generate linear data with gaussian noise for linear regression 
```

```python
rng = np.random.default_rng(seed=42)
m = 200 #number of instances
x = 2 * rng.random((m,1)) #column matrix
y = 4 + 3 * x + rng.standard_normal((m,1))    #i.e. y = 4 + 3x + gaussian noise 
```

```python
#predicting optimal parameters using the normal equation
```

<b><h4>Linear reg equation: <h4></b><br>
$ \hat\theta = \theta^{T} X $<br>             
<b><h4>Normal Equation for finding optimal parameter:</h4></b><br>
$ \theta = (X^{T}X)^{-1}X^{T}y $            

```python
#representation of our data
```

```python
plt.scatter(x , y , s=5 , c="b")
plt.grid()
plt.axis([0.,2.,0,14])
```

```python
x_b = add_dummy_feature(x)
```

```python
x_b.shape
```

```python
x_b
```

```python
x
```

```python
print(f"shape of original {x.shape} , shape of new {x_b.shape}")
```

```python
# we add dummy data for bias
```

```python
theta_best = np.linalg.inv(x_b.T @ x_b) @ x_b.T @ y

theta_best
```

```python
x_new = np.array([[0],
                 [2]])

# alt np.array([[0] , [2]])
```

```python
x_new.shape
```

```python
x_new_b = add_dummy_feature(x_new)
```

```python
y_predict = x_new_b @ theta_best
```

```python
y_predict
```

```python
plt.scatter(x , y , s=5 , c="b")
plt.plot(x_new , y_predict , "r-" , label="Prediction Using Normal Equatino")
plt.legend(loc="upper left")
plt.axis([0.,2.,0,14])
plt.title("Linear Reg parameter using Normal Equation")
plt.grid()
plt.figure(figsize=(10,10))
plt.show()
```
